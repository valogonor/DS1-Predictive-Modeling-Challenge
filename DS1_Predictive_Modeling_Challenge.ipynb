```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
pd.set_option('max_columns', 100)
train = pd.read_csv('https://raw.githubusercontent.com/valogonor/DS1-Predictive-Modeling-Challenge/master/train_features.csv')
test = pd.read_csv('https://raw.githubusercontent.com/valogonor/DS1-Predictive-Modeling-Challenge/master/test_features.csv')  # Unlabeled, for Kaggle submission
labels = pd.read_csv('https://raw.githubusercontent.com/valogonor/DS1-Predictive-Modeling-Challenge/master/train_labels.csv')
train = train.join(labels.status_group, how='inner')
train = train.drop('id', axis=1)
train.head()
```

```
train.isna().sum()
```

```
train_no_nulls = train.dropna(axis=1)
train_no_nulls.head()
```

```
cols = ['recorded_by', 'lga', 'ward', 'wpt_name']
train = train_no_nulls.drop(columns=cols)
train.head()
```

```
train['date_recorded'].dtype
```

```
train.date_recorded = pd.to_datetime(train.date_recorded)
train['date_recorded'].dtype
```

```
train['year_recorded'] = pd.DatetimeIndex(train.date_recorded).year
train.head()
```

```
train.construction_year.value_counts()
```

```
train.describe()
```

```
import numpy as np
train.construction_year = train.construction_year.replace(0, np.nan)
train.describe()
```

```
train.num_private.value_counts()
```

Since almost all of num_private is zero, I'm going to drop this column.

```
train = train.drop('num_private', axis=1)
train.describe()
```

```
train.year_recorded.value_counts()
```

```
train['age'] = train.year_recorded - train.construction_year
train.describe()
```

```
train.loc[train['age'] < 0]
```

Some of the ages are negative. Probably something going on there, but since there are only a few such observations, I'm just going to leave it as is.

```
train.construction_year = train.construction_year.replace(np.nan, 2000.0)
train.age = train.year_recorded - train.construction_year
train.describe()
```

```
train.loc[train['age'] < 0]
```

Even after filling nans in construction year with the median year (2000), it's still just the same 9 observations that have a negative age.

```
# Now to add columns to the test set to match the training set.
test.date_recorded = pd.to_datetime(test.date_recorded)
test['year_recorded'] = pd.DatetimeIndex(test.date_recorded).year
test['age'] = test.year_recorded - test.construction_year
test.head()
```

```
# Splitting the training set so we can validate the model
X = train.drop(['status_group', 'date_recorded'], axis=1) #Dropping date_recorded as it has been replaced with year_recorded
y = train.status_group
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```
# Creating, fitting, nd testing the model
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score

pipeline = make_pipeline(
    ce.OneHotEncoder(use_cat_names=True), 
    StandardScaler(), 
    LogisticRegression(solver='liblinear', multi_class='ovr')
)

pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)
pipeline.score(X, y)
```

```
accuracy_score(y_test, y_pred)
```

```
# Now to drop columns from the test set to match the training set.
test_no_nulls = test.dropna(axis=1)
cols = ['recorded_by', 'lga', 'ward', 'wpt_name', 'date_recorded', 'num_private']
test = test_no_nulls.drop(columns=cols)
test.columns
```

```
test['status_group'] = pipeline.predict(test.drop(columns='id'))
```
