```python
import pandas as pd
pd.set_option('max_columns', 100)
train = pd.read_csv('train_features.csv.zip')
test = pd.read_csv('test_features.csv.zip')  # Unlabeled, for Kaggle submission
labels = pd.read_csv('train_labels.csv.zip')
train = train.join(labels.status_group, how='inner')
train.head()
```

```python
train.describe()
```

```python
train.status_group.value_counts()
```

```python
id = train.id
id.head()
```

```python
train = train.drop('id', axis=1)
train.head()
```

```python
train.dtypes
```

```python
train.date_recorded = pd.to_datetime(train.date_recorded)
train.dtypes
```

```python
from sklearn.model_selection import train_test_split
train_num = train.select_dtypes('number')
X = train_num
y = train.status_group
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression().fit(X_train, y_train)
log_reg.score(X, y)
```

```python
from sklearn.metrics import accuracy_score, classification_report
y_pred = log_reg.predict(X_test)
print(classification_report(y_test, y_pred))
print('accuracy', accuracy_score(y_test, y_pred))
```
